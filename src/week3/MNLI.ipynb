# MNLI 데이터셋
> **Multi-genre Natural Langual Inference** - 다중 장르 자연어 추론   

[![MNLI](https://img.shields.io/badge/Kaggle-MNLI-blue?logo=kaggle)](https://www.kaggle.com/datasets/thedevastator/unlocking-language-understanding-with-the-multin)

- `premise`: **기준**이 되는 문장
- `hypothesis`: **가정**이 되는 문장

### Task

``` plainText
premise와 hypothesis 두 문장 간 논리적 관계를 추론(분류)하는 태스크
```
### Label
- Entailment(함의)
```
premise ✅ => hypothesis ✅
```
- Neutral(중립)
```
논리적 연관 ❌ / 모순도 아닌 관계
```
- Contradiction(모순)
```
premise ✅ => hypothesis ❌
```

<br>

### Goal
```
👉🏻 문장 간 관계 추론(NLU) 대표 태스크 / 문맥 이해와 추론 능력에 비중
```
